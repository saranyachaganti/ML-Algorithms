### Importing necessary libraries

```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn import linear_model, preprocessing
```

### Reading the data

```python
data = pd.read_csv("car.data")
print(data.head())
```

### Convert non numeric to numeric

```python
le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))
```

```python
print(buying)
```

```python
predict = "class"
```

```python
X = list(zip(buying, maint, door, lug_boot, safety))
Y = list(cls)
```

### Splitting test and train

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.1)
```

### Fitting to the model

```python
# you can change the n_neighbors andobserve the results
model = KNeighborsClassifier(n_neighbors = 3)
```

```python
model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print("The accuracy of the model is" ,acc)
```

```python
predicted = model.predict(x_test)
names = ["unacc", "acc", "good", "vgood"]
```

### How much the distance is from the other neighbor (optional)

```python
for x in range(len(predicted)):
    print("Predicted:", predicted[x], "Data: ", x_test[x], "Actual: ", y_test[x], names[y_test[x]])
    n = model.kneighbors([x_test[x]], 9 ,True)
    print(" :N", n)
```
